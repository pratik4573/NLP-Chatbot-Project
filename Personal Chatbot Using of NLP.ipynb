```python
# AI Chatbot with NLP (Jupyter Notebook)


#Step 1: Import Libraries
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
import joblib
import random
```

```python
# Step 3: Train Model
X = df['text']
y = df['intent']

vectorizer = TfidfVectorizer()
X_vec = vectorizer.fit_transform(X)

clf = LogisticRegression()
clf.fit(X_vec, y)

# Save model and vectorizer
joblib.dump(clf, "model.pkl")
joblib.dump(vectorizer, "vectorizer.pkl")

print("âœ… Model trained and saved successfully!")
```

```python
streamlit_code = """
import streamlit as st
import joblib, random
import pandas as pd

# Load model and vectorizer
clf = joblib.load("model.pkl")
vectorizer = joblib.load("vectorizer.pkl")


# Example responses (you can expand this based on your dataset)
responses = {
    "greet": ["Hello! ðŸ‘‹", "Hi there!", "Hey!"],
    "goodbye": ["Goodbye ðŸ‘‹", "See you later!", "Bye!"],
    "wellbeing": ["I'm doing great, thanks!", "All good, how about you?"],
    "name": ["I'm your friendly chatbot ðŸ¤–", "You can call me Chatbot!"],
    "identity": ["You are a valued user ðŸ˜Š", "You're Pratik, right?"]
}

st.title("ðŸ¤– NLP Chatbot")

user_input = st.text_input("You:", "")

if st.button("Send"):
    if user_input:
        X = vectorizer.transform([user_input])
        intent = clf.predict(X)[0]
        if intent in responses:
            st.write("Bot:", random.choice(responses[intent]))
        else:
            st.write("Bot: Sorry, I didn't understand that ðŸ˜…")
"""

# âœ… Use UTF-8 encoding to avoid UnicodeEncodeError
with open("app_streamlit.py", "w", encoding="utf-8") as f:
    f.write(streamlit_code)

print("âœ… Streamlit app file created! Run: streamlit run app_streamlit.py")
```
